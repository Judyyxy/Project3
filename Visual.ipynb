```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib.patches as mpatches
```

```python
df=pd.read_csv('results.csv')
```

```python
df.head()
```

```python
df2=df.pivot(index='hour',columns='name',values='_col2')
df2.head()
```

```python
for i in df2.columns:
    df2[i].plot()

plt.title('Price Evolution Comparison')

plt.xticks(rotation=70)
plt.legend(df2.columns)
```

```python
fig, axs = plt.subplots(10,figsize=(5,8),sharex=True)
fig.suptitle('Vertically Stacked Stock Prices During the Day')

x=0
l=[]
line_labels = list(df2.columns) 
color=["red","orange","gray",""]

for col in df2.columns:
    l.append(axs[x].plot(df2[col])[0])
    axs[x].legend(l[x],labels=[line_labels[x]],loc="best",borderaxespad=0.1)
  
    x=x+1

plt.show()
```
